```python
import numpy as np
embeddings = np.load("../models/document_embeddings.npy")
```

```python
#reduce dimensions using PCA

from sklearn.decomposition import PCA
pca = PCA(n_components=2)

reduced = pca.fit_transform(embeddings)
```

```python
#cluster labels
import joblib
gmm = joblib.load("../models/gmm_cluster_model.pkl")
clusters = gmm.predict(embeddings)
```

```python
#plotting
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

plt.scatter(
    reduced[:, 0], reduced[:, 1], 
    c=clusters, cmap='tab20', s=5
)

plt.title("Document cluster visuallization")
plt.show()
```
